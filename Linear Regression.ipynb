```python
from sklearn.cross_validation import train_test_split
```

```python
x2=np.array(x2)
y=np.array(y)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x2,y,test_size=0.2)
```

```python
len(xtest)/len(xtrain)
```

```python
xtrain
```

```python
#hypothesis function
def h(a,x,b):
    return (a*x)+b
#error function
def error(a,x,b,y):
    e=0
    m=len(y)
    for i in range(m):
        e+=np.power(h(a,x[i],b)-y[i],2)
    return e/(2*m)
#step gradient function --gradient descent
def step_gradient(a,b,x,y,learning_rate):
    grad_a=0
    grad_b=0
    m=len(y)
    for i in range(m):
        grad_a+=(2/m)*(h(a,x[i],b)-y[i])*x[i]
        grad_b+=(2/m)*(h(a,x[i],b)-y[i])
    a=a-(grad_a*learning_rate)
    b=b-(grad_b*learning_rate)
    return a,b

errors=[]

def descend(init_a,x,init_b,y,learning_rate,iterations):
    a=init_a
    b=init_b
    for i in range (iterations):
        a,b=step_gradient(a,b,x,y,learning_rate)
        
        e=error(a,x,b,y)
        errors.append(e)

        if i%1000==0:
            print(e)
            
    return a,b       
```

```python
init_a=0
init_b=0
learning_rate=0.01
iterations=10000
```

```python
final_a,final_b=descend(init_a,xtrain,init_b,ytrain,learning_rate,iterations)
```

```python
plt.plot(xtrain,ytrain,'r.',
        xtrain,h(init_a,xtrain,init_b),'b',
        xtrain,h(final_a,xtrain,final_b),'g')
```

```python
error(final_a,xtrain,final_b,ytrain)
```
